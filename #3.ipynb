## Variable Length Function Arguments

```python
def f(x):
    print(x)
```

```python
f('Hello')
```

```python
f('Hello', 'World')
```

```python
def Argv(*argv):
    for i in argv:
        print(i)
```

```python
Argv('Hello', 'World', 'I', 'am', 'hungry')
```

```python
def ARGV1(arg1, *argv):
    print(arg1)
    print(' ')
    for i in argv:
        print(i)
```

```python
ARGV1('Hello', 'World', 'I', 'am', 'hungry')
```

```python
def ARGV2(*argv, arg1):
    print(arg1)
    print(' ')
    for i in argv:
        print(i)
```

```python
ARGV2('Hello', 'World', 'I', 'am', 'hungry')
```

```python
ARGV2('Hello', 'World', 'I', 'am', 'hungry', arg1 = 'Ya')
```

## Variable Length Keyword Argument

```python
def KWARG(**kwargs):
    for key, values in kwargs.items():
        print(f'{key} = {values}')
```

```python
KWARG(first = 'N', second = 'T', third = 'U')
```

```python
def Four(arg1, arg2, arg3, arg4, arg5):
    print('arg1 is ', arg1)
    print('arg2 is ', arg2)
    print('arg3 is ', arg3)
    print('arg4 is ', arg4)
    print('arg5 is ', arg5)
```

```python
args = ['N', 'T', 'U', 'B', 'A']
```

```python
Four(*args)
```

```python
kwargs = {'arg1': 'N', 'arg2': 'T', 'arg3': 'U', 'arg4': 'B', 'arg5': 'A'}
```

```python
Four(**kwargs)
```
