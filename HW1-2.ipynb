```python
%matplotlib inline
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np

subject_list = []
credit_list = []
grade_list = []

while True:
    subject = input("請輸入科目名稱 (按q離開): ")
    if subject == "q":
        break
    subject_list.append(subject)
    credit = input("請輸入學分數: ")
    credit_list.append(credit)
    grade = input("請輸入成績(等第積分制) (按h提供轉換表): ")
    if grade == "h":
        print("A+: 4.3\nA: 4.0\nA-: 3.7\nB+: 3.3\nB: 3.0\nB-: 2.7\nC+: 2.3\nC: 2.0\nC-:1.7\nF: 0.0")
        grade1 = input("請輸入成績(等第積分制): ")
        grade_list.append(grade1)
    else:
        grade_list.append(grade)

npcredit = np.asarray(credit_list, dtype = np.float)
nppoint = np.asarray(grade_list, dtype = np.float)
gpa = sum(npcredit*nppoint)/sum(npcredit)
print("您這學期的GPA是: " , gpa)

plt.subplot(121)
plt.bar(subject_list, nppoint, color = "orange")
plt.title("Grade Bar Chart")
plt.subplot(122)
plt.pie(npcredit, labels = subject_list, autopct='%1.1f%%')
plt.title("Credit Pie Chart")
plt.axis('equal')
plt.show()
```
