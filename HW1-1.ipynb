```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, np.pi, 1000)
x = np.sin(t)
y = np.cos(t) + np.power(x, 2.0/3)

plt.plot(x, y)
plt.plot(-x, y)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 1000)
x = 2*np.cos(t) - np.cos(2*t)
y = 2*np.sin(t) - np.sin(2*t)

plt.plot(y, x, color = 'r', linewidth = 2)
```
